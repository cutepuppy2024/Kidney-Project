```python
# ID, No 컬럼 삭제
df.drop(["ID","No_Pation"],axis=1,inplace=True)
```

```python
obj_cols = ["Gender","CLASS"]
num_cols = ['AGE', 'Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']
```

```python
import seaborn as sns

sns.set_theme(style="white", context="talk")
plt.figure(figsize =(12,6))

for i, col in enumerate(obj_cols):
    ax = plt.subplot(1,2,i+1)
    sns.countplot(x=df[col],palette='rocket')
    plt.xlabel(col)
plt.show()
```

```python
sns.set_theme(style="white", context = "talk")

plt.figure(figsize=(20,20))

for i, col in enumerate(num_cols):
    ax = plt.subplot(4,3,i+1)
    sns.distplot(df[col])
    plt.xlabel(col)
plt.show()
```

```python
# Creatinine 최대값 800으로, 비정상적으로 높은 수치 보여 이상치 확인

plt.figure(figsize=(5,5))
df.boxplot(column=["Cr"], color="b")

plt.show()
```

```python
# Creatinine 0.5~1.4mg/dL => 400이상의 값에서 존재하는 데이터 삭제

df[df["Cr"]>250] 
```

```python
new_df = df[df["Cr"] < 400] 
new_df
```

```python
# 전체에 대한 대략적인 이상치 확인
plt.figure(figsize=(30,20))
for i, col in enumerate(num_cols):
    ax = plt.subplot(4,3,i+1)
    df.boxplot(column=f"{col}", color="g")

plt.show()
```

```python
# 성별에 다른 당뇨여부

plt.figure(figsize=(14,5))

label = new_df["Gender"].value_counts().index
value = new_df["Gender"].value_counts().values

plt.subplot(1,2,1)
plt.pie(value, labels=label, autopct="%.1f%%", counterclock=True, startangle=90)
plt.title("Gender")

plt.subplot(1,2,2)
sns.countplot(x = df["Gender"], hue=df["CLASS"] ,palette='rocket')
plt.show()
```

```python
# 연령에 따른 당뇨여부

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.histplot(new_df, x="AGE")
plt.xlabel("AGE")

plt.subplot(1,2,2)
sns.boxplot(new_df, x="CLASS",y="AGE", color = "orange")
plt.show()
```

```python
# 당뇨 여부에 따른 Urea, Cr 

plt.figure(figsize=(14,10))

plt.subplot(2,2,1)
sns.histplot(new_df, x="Urea")
plt.xlabel("Urea")

plt.subplot(2,2,2)
sns.boxplot(new_df, x="CLASS",y="Urea")


plt.subplot(2,2,3)
sns.histplot(new_df, x="Cr")
plt.xlabel("Cr")

plt.subplot(2,2,4)
sns.boxplot(new_df, x="CLASS",y="Cr")
plt.show()
```

```python
# 당뇨 여부에 따른 HbA1c

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.histplot(new_df, x="HbA1c")
plt.xlabel("HbA1c")

plt.subplot(1,2,2)
sns.boxplot(new_df, x="CLASS",y="HbA1c", color="orange")
plt.show()
```

```python
# 당뇨 여부에 따른 BMI

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.histplot(new_df, x="BMI")
plt.xlabel("BMI")

plt.subplot(1,2,2)
sns.boxplot(new_df, x="CLASS",y="BMI", color="orange")
plt.show()
```

```python
# 당뇨 여부에 따른 종류별 Cholesterol

plt.figure(figsize=(30,8))

plt.subplot(1,5,1)
sns.boxplot(new_df, x="CLASS",y='Chol')

plt.subplot(1,5,2)
sns.boxplot(new_df, x="CLASS",y='TG')

plt.subplot(1,5,3)
sns.boxplot(new_df, x="CLASS",y='HDL')

plt.subplot(1,5,4)
sns.boxplot(new_df, x="CLASS",y='LDL')

plt.subplot(1,5,5)
sns.boxplot(new_df, x="CLASS",y='VLDL')
plt.show()
```
